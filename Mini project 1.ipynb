```python
xlsxfile_path = "Employee Sample Data.xlsx"
```

```python
dfxlsx = pd.read_excel(xlsxfile_path)
```

```python
dfxlsx = dfxlsx.drop(columns=['EEID', 'Full Name','Job Title','Business Unit','Gender','Ethnicity','Age', 'Hire Date','Bonus %','Country','City','Exit Date' ])
```

```python
Accounting_median = dfxlsx[dfxlsx['Department'] == 'Accounting'].groupby('Department')['Annual Salary'].median()
```

```python
IT_median = dfxlsx[dfxlsx['Department'] == 'IT'].groupby('Department')['Annual Salary'].median()
```

```python
Finance_median = dfxlsx[dfxlsx['Department'] == 'Finance'].groupby('Department')['Annual Salary'].median()
```

```python
Marketing_median = dfxlsx[dfxlsx['Department'] == 'Marketing'].groupby('Department')['Annual Salary'].median()
```

```python
HR_median = dfxlsx[dfxlsx['Department'] == 'Human Resources'].groupby('Department')['Annual Salary'].median()
```

```python
Sales_median = dfxlsx[dfxlsx['Department'] == 'Sales'].groupby('Department')['Annual Salary'].median()
```

```python
Engineering_median = dfxlsx[dfxlsx['Department'] == 'Engineering'].groupby('Department')['Annual Salary'].median()
```

```python
xlsx_data = {'departments':['IT','Accounting','Finance','Marketing','HR','Sales','Engineering'],
            'annual salary':[int(IT_median),int(Accounting_median),int(Finance_median),int(Marketing_median),int(HR_median),int(Sales_median),int(Engineering_median)]}
```

```python
xlsxmedian = pd.DataFrame(xlsx_data)
```

```python
xlsxmedian.plot(x='departments',y='annual salary',kind='bar')
plt.ylabel('Annual salary')
plt.title('Departments median annual salary(in usd)')
```
